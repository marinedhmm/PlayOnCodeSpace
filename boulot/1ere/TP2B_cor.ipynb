```python
# Ex1
def premiereApparition(mot, lettre):
    """
    retourne l'indice de la première apparition (en partant de l'indice 0) de la lettre dans le mot.
    Si la lettre n'apparaît pas alors la fonction retourne -1
    """
    assert type(mot) == str  and type(lettre) == str
    assert len(lettre) == 1
    for i in range(len(mot)):
        if lettre == mot[i]:
            return i
    return -1

#tests
premiereApparition("abracadabra", "a") == 0
premiereApparition("abracadabra", "b") == 1
premiereApparition("abracadabra", "c") == 4
premiereApparition("abracadabra", "e") == -1
print("bvo")
```

```python
# Ex2
def renversement(mot):
    """
    renvoie la chaîne mot à l'envers.
    """
    reponse = ""
    for lettre in mot:
        reponse = lettre + reponse 
    return reponse


#tests
assert renversement('informatique') == 'euqitamrofni'
assert renversement('nsi') == 'isn'
assert renversement('') == ''
print("bvo")
```

```python
# Ex3
def retireVoyelle (mot):
    """
    retourne le "même mot" privé de toutes ses voyelles
    """
    assert type(mot) == str
    voyelles = "aeiouy"
    reponse = ""
    for lettre in mot:
        if not lettre in voyelles:
            reponse += lettre
    return reponse

#tests
assert retireVoyelle ("bonjour") == 'bnjr'
assert retireVoyelle ("salut") == 'slt'

def chiffrement(mot, cle):
    """
    Enlève les voyelles du mot initial (avec la fonction précédente) 
    et les lettres restantes sont transformées de la façon suivante:
        + On récupère le code unicode de chaque lettre 
        + On ajoute cle pour obtenir un nouveau code
        + On récupère la lettre correspondant à ce nouveau code.
    """
    assert type(mot) == str and type(cle) == int
    motSansVoyelle = retireVoyelle (mot)
    reponse = ""
    for lettre in motSansVoyelle:
        code = ord(lettre)
        nouveauCode = code + cle
        reponse = reponse + chr(nouveauCode)
    return reponse

#tests
assert chiffrement("b", 1) == 'c'
assert chiffrement ("bonjour", 1) == 'coks'
assert chiffrement("bonjour", 30000) == '疒疞疚疢' 
print("bvo")

```

```python
# Ex4
def correspond(motif, chaine, position):
    """
    renvoie True si on retrouve motif exactement à partir
    de position dans chaine et retourne False sinon. 
    """
    assert type(motif) == str and type(chaine) == str
    assert type(position) == int
    for i in range(len(motif)):
        if motif[i] != chaine[position + i]:
            return False
    return True


#tests
assert not correspond("AA", "AAGGTTCC", 4) 
assert correspond("AT", "ACGCATGC", 4)
print("bvo")
```
